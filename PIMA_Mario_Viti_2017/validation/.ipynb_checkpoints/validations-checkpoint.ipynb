# Validation

```python
import json
from pprint import pprint
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import numpy as np
from os import listdir
from os.path import isfile, join

def rad_to_deg(a):
    return (np.pi/180.0)*a

def cartesian_to_polar((x,y)):
    if np.abs(x) <0.0001:
        x = 0.0001
    if np.abs(y) <0.0001:
        y = 0.0001
    r = np.sqrt(x**2 + y**2)
    angle = rad_to_deg(np.arctan(y/float(x)))
    return r,angle

mypath = 'validations'

onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
db = {}
for afile in onlyfiles:
    name = afile[:-5]
    with open(join(mypath,afile)) as data_file:
        data = json.load(data_file)
        db[name] = data
    
size = 8
full, low, rot = np.zeros((size,size)),np.zeros((size,size)),np.zeros((size,size))
full_err, low_err, rot_err = np.zeros((size,size)),np.zeros((size,size)),np.zeros((size,size))
full_err_x,full_err_y,full_dev,full_angle = [],[],[],[]
low_err_x,low_err_y,low_dev,low_angle = [],[],[],[]
rot_err_x,rot_err_y,rot_dev,rot_angle = [],[],[],[]
err_x,err_y,err_dev,err_angle = [],[],[],[]

max_x,min_x,max_y,min_y = 0,0,0,0
for k in db.keys():
    tags_table = db[k]['tags']
    for tag_k in tags_table:
        x,y = tags_table[tag_k]['pos']
        x,y = int(x), int(y)
        max_x = x if x > max_x else max_x
        min_x = x if x < min_x else min_x
        max_y = y if y > max_y else max_y
        min_y = y if y < min_y else min_y
    max_x += 1.001
    max_y += 1.001
    for tag_k in tags_table:
        x,y = tags_table[tag_k]['pos']
        x,y = int(x), int(y)
        ro,theta = cartesian_to_polar((x,y))
        err_angle += [theta]
        u,v = np.abs(x)/float(max_x),  np.abs(y)/float(max_y)
        err_x +=  [tags_table[tag_k]['dist']]
        err_y += [tags_table[tag_k]['abs_err_d']]
        err_dev += [tags_table[tag_k]['std_d']]
        if 'rot' in k:
            rot_angle += [theta]
            h,w = rot.shape
            px = int(np.floor(u*w))
            py = int(np.floor(v*h))
            rot[px,py] = tags_table[tag_k]['hit_rel_freq']
            rot_err[px,py] = tags_table[tag_k]['abs_err_d']
            rot_err_x += [tags_table[tag_k]['dist']]
            rot_err_y += [tags_table[tag_k]['abs_err_d']]
            rot_dev += [tags_table[tag_k]['std_d']]
        elif'full' in k:
            full_angle += [theta]
            h,w = full.shape
            px = int(np.floor(u*w))
            py = int(np.floor(v*h))
            full[px,py] = tags_table[tag_k]['hit_rel_freq']
            full_err[px,py] = tags_table[tag_k]['abs_err_d']
            full_err_x += [tags_table[tag_k]['dist']]
            full_err_y += [tags_table[tag_k]['abs_err_d']]
            full_dev += [tags_table[tag_k]['std_d']]
        elif 'low' in k:
            low_angle += [theta]
            h,w = low.shape
            px = int(np.floor(u*w))
            py = int(np.floor(v*h))
            low[px,py] = tags_table[tag_k]['hit_rel_freq']
            low_err[px,py] = tags_table[tag_k]['abs_err_d']
            low_err_x += [tags_table[tag_k]['dist']]
            low_err_y += [tags_table[tag_k]['abs_err_d']]
            low_dev += [tags_table[tag_k]['std_d']]
            
            
err_angle_h = zip(err_angle,err_y,err_dev)
err_angle_h = np.array(sorted(err_angle_h, key=lambda tup: tup[0])).reshape(len(err_angle),3)
full_angle_h = zip(full_angle,err_y,err_dev)
full_angle_h = np.array(sorted(full_angle_h, key=lambda tup: tup[0])).reshape(len(full_angle),3)
low_angle_h = zip(low_angle,err_y,err_dev)  
low_angle_h = np.array(sorted(low_angle_h, key=lambda tup: tup[0])).reshape(len(low_angle),3)
full_angle_h = zip(full_angle,err_y,err_dev) 
full_angle_h = np.array(sorted(full_angle_h, key=lambda tup: tup[0])).reshape(len(full_angle),3)

err_h = zip(err_x,err_y,err_dev)            
err_h = np.array(sorted(err_h, key=lambda tup: tup[0])).reshape(len(err_x),3)
full_err_h = zip(full_err_x,full_err_y,full_dev)            
full_err_h = np.array(sorted(full_err_h, key=lambda tup: tup[0])).reshape(len(full_err_x),3)
low_err_h = zip(low_err_x,low_err_y,low_dev)
low_err_h = np.array(sorted(low_err_h, key=lambda tup: tup[0])).reshape(len(low_err_x),3)
rot_err_h = zip(rot_err_x,rot_err_y,rot_dev)            
rot_err_h = np.array(sorted(rot_err_h, key=lambda tup: tup[0])).reshape(len(rot_err_x),3)

full = np.vstack((np.flipud(full),full))
full_err = np.vstack((np.flipud(full_err),full_err))
full, full_err = np.fliplr(full).T,  np.fliplr(full_err).T

low = np.vstack((np.flipud(low),low))
low_err = np.vstack((np.flipud(low_err),low_err))
low,low_err = np.fliplr(low).T,  np.fliplr(low_err).T

rot = np.vstack((np.flipud(rot),rot))
rot_err = np.vstack((np.flipud(rot_err),rot_err))
rot,rot_err = np.fliplr(rot).T,  np.fliplr(rot_err).T
```

# Measuring error and frequence of detection.

Experimental validation for vision system of the ThymioPYPI framework.
* evaluating frequence of detection.
* evaluationg distance estimation vs distance and angle.
* map of the frequence of detection.
* map of the absolute error of the estimated distance and the actual distance.

## Frequence of detection.

Maximum frequence is around $f_{max} = 12 hz$, frequece of detection is the ration of the number of tag found per second $f_{detected}$ wile running and the maximum value. $$ f_{detection} = \frac{f_{detected}}{f_{max}}  $$


## Settings of detection.

measuerment has been taken in 3 different settings, shown in the images below.

```python
path = "pics"
onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]
f, axarr = plt.subplots(2,len(onlyfiles)/2,figsize=(15,15))
i = 0
for afile in onlyfiles:
    img=mpimg.imread(join(path,afile))
    axarr[i/2,i%2].set_title(afile[:-4])
    axarr[i/2,i%2].imshow(img)
    i += 1
plt.show()
```

```python

f, axarr = plt.subplots(1,2,figsize=(10,3))

x,y,e = full_angle_h[:,0],full_angle_h[:,1],full_angle_h[:,2]
axarr[0].set_title("full light")
axarr[0].errorbar(x, y, e/float(2), linestyle='None', fmt='-o')
plt.xlabel('angle')
plt.ylabel('abs_error')

x,y,e = err_angle_h[:,0],err_angle_h[:,1],err_angle_h[:,2]
axarr[1].set_title("all settings (full light, low light, rotation)")
axarr[1].errorbar(x, y, e/float(2), linestyle='None', fmt='-o')

plt.xlabel('angle')
plt.ylabel('abs_error')
plt.tight_layout()
plt.show()

f, axarr = plt.subplots(1,2,figsize=(10,3))

x,y,e = full_err_h[:,0],full_err_h[:,1],full_err_h[:,2]
fit = np.polyfit(x,y,2)
fit_fn = np.poly1d(fit) 
axarr[0].set_title("full light")
axarr[0].errorbar(x, y, e/float(2), linestyle='None', fmt='-o')
axarr[0].plot(x, fit_fn(x), '--k')
plt.xlabel('distance cm')
plt.ylabel('abs_error')

x,y,e = err_h[:,0],err_h[:,1],err_h[:,2]
fit = np.polyfit(x,y,2)
fit_fn = np.poly1d(fit) 

axarr[1].set_title("all settings (full light, low light, rotation)")
axarr[1].errorbar(x, y, e/float(2), linestyle='None', fmt='-o')
axarr[1].plot(x, fit_fn(x), '--k')

plt.xlabel('distance cm')
plt.ylabel('abs_error')
plt.tight_layout() 
plt.show()

interp = 'spline16'
ratio = 1.0
f, axarr = plt.subplots(3, 2,figsize=(22,22))
axarr[0,0].set_title("full light frequence of detection")
axarr[0,0].imshow(full, interpolation=interp, extent=[max_x,-max_x,min_y,max_y], aspect=ratio)
axarr[0,0].grid(True)

axarr[0,1].set_title("full light dist abs err")
axarr[0,1].imshow(full_err, interpolation=interp, extent=[max_x,-max_x,min_y,max_y], aspect=ratio)
axarr[0,1].grid(True)

axarr[1,0].set_title("rotated in full light frequence of detection")
axarr[1,0].imshow(rot, interpolation=interp, extent=[max_x,-max_x,min_y,max_y], aspect=ratio)
axarr[1,0].grid(True)

axarr[1,1].set_title("rotated in full light dist abs err")
axarr[1,1].imshow(rot_err, interpolation=interp, extent=[max_x,-max_x,min_y,max_y], aspect=ratio)
axarr[1,1].grid(True)

axarr[2,0].set_title("low light frequence of detection")
axarr[2,0].imshow(low, interpolation=interp, extent=[max_x,-max_x,min_y,max_y], aspect=ratio)
axarr[2,0].grid(True)

axarr[2,1].set_title("rotated in full light dist abs err")
axarr[2,1].imshow(low_err, interpolation=interp, extent=[max_x,-max_x,min_y,max_y], aspect=ratio)
axarr[2,1].grid(True)

plt.tight_layout() 
plt.grid(True)
plt.show()
```
